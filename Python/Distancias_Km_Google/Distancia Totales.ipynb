```python
import pandas as pd

# Cargar el archivo Excel
df = pd.read_excel('REPORTE_google_con_distancia.xlsx')

# 1. Filtrar registros donde PERIODO sea 3
df = df[df['PERIODO'] == 3]

# 2. Filtrar solo los medios de transporte específicos (4, 5, 6)
df = df[df['P7'].isin([4, 5, 6])]

# 3. Mapear valores en la columna P1
def mapear_p1(valor):
    if valor == 1:
        return 2
    elif valor == 2 or valor == 3:
        return 4
    else:
        return valor

df['P1'] = df['P1'].apply(mapear_p1)

# 4. Calcular distancia total por medio de transporte
df['Distancia_Total'] = df['DISTANCIA_GOOGLE_KM'] * df['P1']

# Agrupar por medio de transporte y calcular totales
distancia_por_transporte = df.groupby('P7')['Distancia_Total'].sum()

# Mapear los códigos a nombres descriptivos
nombres_transporte = {
    4: 'Chofer',
    5: 'Acompañante', 
    6: 'Bus'
}

print("DISTANCIA TOTAL RECORRIDA POR MEDIO DE TRANSPORTE (PERIODO 3)")
print("=" * 60)

for codigo, distancia in distancia_por_transporte.items():
    nombre = nombres_transporte.get(codigo, f'Código {codigo}')
    print(f"{nombre}: {distancia:.2f} km")

# Total general
total_general = distancia_por_transporte.sum()
print("=" * 60)
print(f"TOTAL GENERAL: {total_general:.2f} km")

# Opcional: Mostrar cantidad de personas por medio de transporte
print("\nCANTIDAD DE PERSONAS POR MEDIO DE TRANSPORTE")
print("=" * 45)
conteo_por_transporte = df.groupby('P7').size()
for codigo, cantidad in conteo_por_transporte.items():
    nombre = nombres_transporte.get(codigo, f'Código {codigo}')
    print(f"{nombre}: {cantidad} personas")
```

```python
import pandas as pd

# Cargar el archivo Excel
ruta = r"C:\Users\Christian\Desktop\TrabajoUDA\Trabajo Victor\Goolgle\REPORTE_google_con_distancia.xlsx"
df = pd.read_excel(ruta)

# --- 1. Filtrar solo PERIODO = 3 ---
df = df[df['PERIODO'] == 3].copy()

print("=== FILTRO DE PERIODO ===")
print(f"Estudiantes con PERIODO = 3: {len(df)}\n")

# --- 2. Mapas de transporte ---
mapa_P7 = {
    1: "Caminando",
    2: "Bicicleta privada",
    3: "Bicicleta pública",
    4: "Auto (como conductor)",
    5: "Auto (como acompañante)",
    6: "Bus",
    7: "Tranvía",
    8: "Taxi",
    9: "Motocicleta",
    10: "Scooter"
}

mapa_P10 = {
    1: "Auto (como acompañante)",
    2: "Auto (como conductor)",
    3: "Bicicleta privada",
    4: "Bicicleta pública",
    5: "Bus",
    6: "Caminando",
    7: "Motocicleta",
    8: "Scooter",
    9: "Taxi",
    10: "Tranvía",
    11: "Otro"
}

df['P7'] = df['P7'].map(mapa_P7)
df['P10'] = df['P10'].map(mapa_P10)

# --- 3. Modificar P1 ---
df['P1'] = df['P1'].replace({1: 2, 2: 4, 3: 4})

# --- 4. Calcular la distancia real ---
df['DISTANCIA_REAL'] = df['P1'] * df['DISTANCIA_GOOGLE_KM']

# --- 5. Definir los pares de cambio ---
cambios = [
    ("Auto (como conductor)", "Auto (como acompañante)"),
    ("Auto (como conductor)", "Bus"),
    ("Bus", "Auto (como conductor)")
]

# Total de estudiantes que actualmente usan Auto (como conductor)
total_auto_conductor = (df['P7'] == "Auto (como conductor)").sum()

print("=== Resultados de cambios de medio de transporte ===\n")

for origen, destino in cambios:
    filtro = (df['P7'] == origen) & (df['P10'] == destino)
    cantidad = filtro.sum()
    distancia_total = df.loc[filtro, 'DISTANCIA_REAL'].sum()
    
    if origen == "Auto (como conductor)":
        porcentaje = (cantidad / total_auto_conductor * 100) if total_auto_conductor > 0 else 0
        print(f"Filtro aplicado: {origen} → {destino}")
        print(f"  Registros filtrados: {cantidad}")
        print(f"  Distancia total (km): {distancia_total:.2f}")
        print(f"  Porcentaje de los que conducían auto: {porcentaje:.2f}%")
    else:
        print(f"Filtro aplicado: {origen} → {destino}")
        print(f"  Registros filtrados: {cantidad}")
        print(f"  Distancia total (km): {distancia_total:.2f}")
    
    print("-" * 60)
```
